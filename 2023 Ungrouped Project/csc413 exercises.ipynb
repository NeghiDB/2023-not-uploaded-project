```python
import random
import csv
from itertools import chain

store_list = []
header = ['name','questions','answers','score']

for i in range (0,2):
    random_no = random.randint(0,9)
    store_list.append(random_no)

name = input('Enter your name ')
correct_answer = [5,6,8,9,1,2,3,4,6,20]

questions = []
answer = []
scored = []
final_score = 0

for i in store_list:
    if i == 0:
        valla = int(input('Find the root of 25'))
        questions.append('Find the root of 25')
        answer.append(valla)

        if valla == correct_answer[0]:
            scored.append(50)
    
    if i == 1:
        valla = int(input('Find the root of 250'))
        questions.append('Find the root of 250')
        answer.append(valla)

        if valla == correct_answer[1]:
            scored.append(50)

for value in scored:
    final_score = final_score + int(value)

print(name)
conver = str(final_score)
print(conver)
s = [name, questions, answer, final_score]

with open('test.csv','a+') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(s)

print(s)
```

```python
import matplotlib.pyplot as plt
a = ['python','java','c','c++','R','Javascript','C#']
popularity = [100,96.3,94.4,87.5,81.5,79.4,74.5]

a_pos = [x for x, x in enumerate(a)]

plt.bar(a_pos, popularity, color='blue', edgecolor='green')

plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("Popularity of programming languages 2019 according to IEEE")
plt.xticks(a_pos, a)

#turns on grid
plt.minorticks_on()
plt.grid(which = 'major', linestyle = '-', linewidth = '0.5', color = 'gray')

#customize the minor grid
plt.grid(which = 'minor', linestyle = ':', linewidth = '0.5', color = 'black')
plt.show()
```
